## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

## Research Question

### Search Question
How can search ranking signals be used to improve the ordering of search results for a user query?

### Decision
The decision is which pages or search results should appear higher in the search results.

### Unit of Analysis
One search result (page) for a search query.

### Output
A ranking score that estimates how relevant a search result is.

### Action
The search system can rank more relevant pages higher and less relevant pages lower.

### Cost of a Wrong Recommendation
If irrelevant pages appear at the top, users may not find what they are looking for. This can reduce user satisfaction, lower engagement, and decrease trust in the search system.

### Why Machine Learning?
Simple rules cannot capture all user behaviors. Machine learning can learn patterns from historical search data and make better ranking decisions.
